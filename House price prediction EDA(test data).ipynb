# Exploratory data analysis for test data.

```python
#visualizing data for test data.
```

```python
#import libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.pandas.set_option('display.max_columns',None)
```

```python
data_frame_inp=pd.read_csv('test.csv')
data_frame_inp.head()
```

```python
data_frame_oup=pd.read_csv('sample_submission.csv')
data_frame_oup.head()
```

```python
data_frame=data_frame_oup.merge(data_frame_inp)
```

```python
data_frame.head()
```

```python
#checking out missing values.
null_value_features=[]
for cols in data_frame.columns:
    percentage=(data_frame[cols].isnull().sum()/len(data_frame))*100
    if percentage>0:
        print(cols,' has ',percentage,'% of null values.')
        null_value_features.append(cols)
```

```python
data_frame[null_value_features].head()
```

```python
#impact of null values on saleprice. 
```

```python
for cols in null_value_features:
    print(cols)
```

```python
for cols in null_value_features:
    data=data_frame.copy()
    data[cols]=np.where(data[cols].isnull(),1,0)
    data.groupby(cols)['SalePrice'].median().plot.bar()
    plt.ylabel('SalePrice')
    plt.show()
```

```python
#there's a Id feature in the dataframe.
```

```python
num_features=[]
for cols in data_frame.columns:
    if data_frame[cols].dtype!='O':
        num_features.append(cols)
```

```python
data_frame[num_features].head()
```

```python
temporal_features=[]
for cols in num_features:
    if 'Year' in cols:
        temporal_features.append(cols)
    elif 'Yr' in cols:
        temporal_features.append(cols)
```

```python
temporal_features
```

```python
for cols in temporal_features:
    print(cols,': ',data_frame[cols].unique())
```

```python
#visualizing YrSold feature.
data_frame.groupby('YrSold')['SalePrice'].median().plot.line()
```

```python
#comparing the temporal features with SalePrice.
```

```python
for cols in temporal_features:
    if cols!='YrSold':
        data=data_frame.copy()
        data[cols]=data['YrSold']-data[cols]
        plt.scatter(data[cols],data['SalePrice'])
        plt.xlabel(cols)
        plt.ylabel('SalePrice')
        plt.show()
```

# Numerical features

```python
numerical_features=[]
for cols in data_frame.columns:
    if data_frame[cols].dtype!='O':
        numerical_features.append(cols)
```

```python
data_frame[numerical_features].head()
```

```python
discrete_variables=[]
continous_variables=[]
for cols in numerical_features:
    if len(data_frame[cols].unique())<25:
        discrete_variables.append(cols)
    else:
        continous_variables.append(cols)
```

```python
continous_variables
```

```python
data_frame[discrete_variables].head()
```

```python
#visualizing discreate_variables.
for cols in discrete_variables:
    data_frame.groupby(cols)['SalePrice'].median().plot.bar()
    plt.ylabel('SalePrice')
    plt.xlabel(cols)
    plt.show()
```

```python
#visualizing continous_variables.
for col in continous_variables:
    if col!='Id':
        plt.hist(data_frame[col])
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()
```

```python
#seeing the continous features with logarithimic values.
for cols in continous_variables:
    data=data_frame.copy()
    if 0 in data[cols].unique():
        pass
    else:
        if cols!='Id':
            data[cols]=np.log(data[cols])
            data[cols].hist()
            plt.xlabel(cols)
            plt.ylabel('SalePrice')
            plt.show()
```

# Outliers

```python
#finding the outliers in the continous features.
for cols in continous_variables:
    plt.boxplot(data_frame[cols])
    plt.xlabel(cols)
    plt.ylabel('Count')
    plt.show()
```

# Categorical features

```python
#getting the categorical features from the dataframe.
categorical_features=[]
for cols in data_frame.columns:
    if data_frame[cols].dtype=='O':
        categorical_features.append(cols)
```

```python
#finding the cardinality of the features(number of categories in the categorical features).
for cols in categorical_features:
    print(cols,' has ',len(data_frame[cols].unique()),' categories')
```

```python
#getting the relationship between the categorical features and output feature.
for cols in categorical_features:
    data_frame.groupby(cols)['SalePrice'].median().plot.bar()
    plt.xlabel(cols)
    plt.ylabel('SalePrice')
    plt.show()
```

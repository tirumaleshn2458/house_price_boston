# Feature engineering

1. Missing values.
2. Temporal variables.
3. Handling numerical variables.
4. Handling categorical variables. Finding Rare_var variables.

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.pandas.set_option('display.max_columns',None)
```

```python
data_frame=pd.read_csv('test_data.csv')
data_frame.head()
```

### Temporal variables.

```python
temporal_var=[]
for cols in data_frame.columns:
    if 'Yr' in cols:
        temporal_var.append(cols)
    elif 'Year' in cols:
            temporal_var.append(cols)
```

```python
temporal_var
```

```python
for cols in temporal_var:
    if cols!='YrSold':
        #getting the difference from YrSold feature because the dependent feature reduced as moving forward.
        data_frame[cols]=data_frame['YrSold']-data_frame[cols]
```

```python
data_frame[temporal_var].head(30)
```

## Missing values.

```python
#finding numeircal and categorical features.
categorical_features=[]
numerical_features=[]
for cols in data_frame.columns:
    if data_frame[cols].dtype=='O':
        categorical_features.append(cols)
    else:
        numerical_features.append(cols)
```

```python
#handling missing values for categorical features.
#finding the percentage of null values.
for cols in categorical_features:
    null_percentage=(data_frame[cols].isnull().sum()/len(data_frame))*100
    if null_percentage>0:
        print(cols,' has ',null_percentage,'% of null values.')
```

```python
#replacing nan values with 'missing'.
for cols in categorical_features:
    data_frame[cols].fillna('Missing',inplace=True)
```

```python
data_frame[categorical_features].head(50)
```

```python
#handling null values for numerical features.
```

```python
#finding the percentage of null values in numerical_features.
null_value_features=[]
for cols in numerical_features:
    percentage=(data_frame[cols].isnull().sum()/len(data_frame))*100
    if percentage>0:
        print(cols,' has ',percentage,'% of null values.')
        null_value_features.append(cols)
```

```python
#replacing null values in numerical features with the median of the feature.
for cols in null_value_features:
    data_frame[cols+'_null']=np.where(data_frame[cols].isnull(),1,0)
    median_value=data_frame[cols].median()
    data_frame[cols].fillna(median_value,inplace=True)
```

```python
for cols in data_frame.columns:
    percentage=(data_frame[cols].isnull().sum()/len(data_frame))*100
    if percentage>0:
        print(cols,' has ',percentage,'% of null values.')
```

```python
#there are no more null features in dataset.
```

```python
data_frame.head(50)
```

### handling categorical features.

```python
#handling rare categorical variables.
for cols in categorical_features:
    percentage=data_frame.groupby(cols)['SalePrice'].count()/len(data_frame)
    percentage_index=percentage[percentage>0.1].index
    data_frame[cols]=np.where(data_frame[cols].isin(percentage_index),data_frame[cols],'Rare_var')
```

```python
data_frame[categorical_features].head(50)
```

```python
#converting categorical features to numerical features
for cols in categorical_features:
    rank=data_frame[cols].value_counts().index
    mapping={i:k for k,i in enumerate(rank,0)}
    data_frame[cols]=data_frame[cols].map(mapping)
```

```python
data_frame.head(20)
```

### Transforming continous data.

```python
continous_data_to_transform=[]
for cols in numerical_features:
    if len(data_frame[cols].value_counts())>25:
        if 0 in data_frame[cols].unique():
            pass
        else:
            continous_data_to_transform.append(cols)
```

```python
for cols in continous_data_to_transform:
    data_frame[cols]=np.log(data_frame[cols])
```

### Feature scaling

```python
features_for_feature_scaling=[feature for feature in data_frame.columns if feature not in ['Id','SalePrice']]
```

```python
new_data=data_frame[features_for_feature_scaling]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(new_data)
```

```python
scaler.transform(new_data)
```

```python
data1=pd.DataFrame(scaler.transform(new_data),columns=new_data.columns)
data2=data_frame[['Id','SalePrice']]
final_data=pd.concat([data1,data2.reset_index(drop=True)],axis=1)
```
